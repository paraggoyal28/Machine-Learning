```python
## RUN IT IN SHELL AS HERE U WILL OT BE ABLE TO SEE TH ELIVE EFFECT

import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style
```

```python
style.use('fivethirtyeight')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    graph_data = open('example.txt','r').read()
    lines = graph_data.split('\n')
    xs = []
    ys = []
    for line in lines:
        if len(line) > 1:
            x,y = line.split(',')
            xs.append(x)       
            ys.append(y)
    ax1.clear() # for animation
    ax1.plot(xs, ys)
    
ani = animation.FuncAnimation(fig, animate, interval=10)  # interval in milisecond
plt.show()
    
```
